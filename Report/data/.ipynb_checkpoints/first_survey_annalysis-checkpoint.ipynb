```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df = pd.read_csv("first_survey.csv", encoding='latin1')
```

```python
## Thematic Annalysis 
import re
from sklearn.feature_extraction.text import CountVectorizer
last_two_columns = df.iloc[:, -2:]
last_two_columns_info = last_two_columns.dtypes
last_two_columns.head(), last_two_columns_info
## data cleaning 
last_two_columns_info = last_two_columns.dtypes
last_two_columns.head(), last_two_columns_info
### cleaning the text 
def clean_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower()
    return text

last_two_columns_cleaned = last_two_columns.applymap(clean_text)
last_two_columns_cleaned.head()


# vectorising
vectorizer = CountVectorizer(stop_words='english')
all_texts = last_two_columns_cleaned['Change_social_media_comments'] + ' ' + last_two_columns_cleaned['Features_to_combat_loneliness']

word_matrix = vectorizer.fit_transform(all_texts)
feature_names = vectorizer.get_feature_names_out()

# sort by frequency
word_counts = word_matrix.sum(axis=0)
word_counts_sorted = sorted(zip(word_counts.tolist()[0], feature_names), reverse=True, key=lambda x: x[0])

### top words 
word_counts_sorted[:20]

revised_theme_groups = {
    "User Experience and Management": ['users', 'user', 'provide', 'option'],
    "Feature Development": ['introduce', 'develop', 'offer', 'create', 'feature'],
    "Comments Management and Removal": ['comment', 'comments', 'delete', 'remove', 'exclude', 'filter', 'mute'],
    "Engagement and Sharing": ['share', 'connect', 'engagement', 'social', 'community']
}

revised_theme_counts = {theme: 0 for theme in revised_theme_groups}

# occourence counting related to the theme
for index, row in last_two_columns_cleaned.iterrows():
    document = row['Change_social_media_comments'] + ' ' + row['Features_to_combat_loneliness']
    words = document.split()
    for word in words:
        for theme, keywords in revised_theme_groups.items():
            if word in keywords:
                revised_theme_counts[theme] += 1

revised_theme_counts
```
